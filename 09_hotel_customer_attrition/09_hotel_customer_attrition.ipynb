# Прогнозирование оттока клиентов в сети отелей «Как в гостях»

# Описание проекта
Заказчик этого исследования — сеть отелей «Как в гостях».<br />
Чтобы привлечь клиентов, эта сеть отелей добавила на свой сайт возможность забронировать номер без предоплаты. Однако если клиент отменял бронирование, то компания терпела убытки. Сотрудники отеля могли, например, закупить продукты к приезду гостя или просто не успеть найти другого клиента.<br />
Чтобы решить эту проблему, вам нужно разработать систему, которая предсказывает отказ от брони. Если модель покажет, что бронь будет отменена, то клиенту предлагается внести депозит. Размер депозита — 80% от стоимости номера за одни сутки и затрат на разовую уборку. Деньги будут списаны со счёта клиента, если он всё же отменит бронь.<br />

# Бизнес-метрика и другие данные
Основная бизнес-метрика для любой сети отелей — её прибыль. Прибыль отеля — это разница между стоимостью номера за все ночи и затраты на обслуживание: как при подготовке номера, так и при проживании постояльца.<br />
В отеле есть несколько типов номеров. В зависимости от типа номера назначается стоимость за одну ночь. Есть также затраты на уборку. Если клиент снял номер надолго, то убираются каждые два дня.<br />
Стоимость номеров отеля:<br />
категория A: за ночь — 1 000, разовое обслуживание — 400;<br />
категория B: за ночь — 800, разовое обслуживание — 350;<br />
категория C: за ночь — 600, разовое обслуживание — 350;<br />
категория D: за ночь — 550, разовое обслуживание — 150;<br />
категория E: за ночь — 500, разовое обслуживание — 150;<br />
категория F: за ночь — 450, разовое обслуживание — 150;<br />
категория G: за ночь — 350, разовое обслуживание — 150.<br />
В ценовой политике отеля используются сезонные коэффициенты: весной и осенью цены повышаются на 20%, летом — на 40%.
Убытки отеля в случае отмены брони номера — это стоимость одной уборки и одной ночи с учётом сезонного коэффициента.
На разработку системы прогнозирования заложен бюджет — 400 000. При этом необходимо учесть, что внедрение модели должно окупиться за тестовый период. Затраты на разработку должны быть меньше той выручки, которую система принесёт компании.

Основная бизнес-метрика для любой сети отелей — её прибыль. Прибыль отеля — это разница между стоимостью номера за все ночи и затраты на обслуживание: как при подготовке номера, так и при проживании постояльца.<br />

```python
profit_elements = df_test.drop(columns = ['lead_time', 'arrival_date_year',
                                          'arrival_date_week_number', 'arrival_date_day_of_month',
                                          'adults', 'children',
                                          'babies', 'meal', 'country', 'distribution_channel',
                                          'is_repeated_guest', 'previous_cancellations',
                                          'previous_bookings_not_canceled', 'booking_changes',
                                          'days_in_waiting_list', 'customer_type', 'required_car_parking_spaces',
                                          'total_of_special_requests'],axis=1)
```

### Корректировка формулы расчета прибыли

Чтобы решить эту проблему, необходимо разработать систему, которая предсказывает отказ от брони. Если модель покажет, что бронь будет отменена, то клиенту предлагается внести депозит. Размер депозита — 80% от стоимости номера за одни сутки и затрат на разовую уборку. Деньги будут списаны со счёта клиента, если он всё же отменит бронь.<br />

Откорректируем написанную ранее функцию для выручки после обучения моделей с внедрением депозитов:<br />
- если система ошибочно предскажет "ненадёжного" клиента и предложит ему предоплату, и он не отказывается от брони и приезжает, то в этом случае доходы и расходы такие же как и если бы система верно предсказывала, что отмены не будет;<br />
- если система ошибочно предскажет, что клиент "надёжный", то отель не предложит ему предоплату, а клиент заказ в итоге отменил, поэтому отель получает убыток в виде стоимости одной уборки и одной ночи с учётом сезонного коэффициента;<br />
- если система верно предскажет "ненадёжного" клиента, то клиенту предложат депозит в размере 80% от стоимости номера за одни сутки и стоимость одной уборки, и если клиент не приедет, тогда в расход пойдет стоимость одной уборки и одной ночи с учётом сезонного коэффициента<br />
- в качестве эксперимента попробуем проверить такой сценарий, назовем его худшим: пусть половина случайных клиентов из тех, с кого модель предложит взять депозит(вне зависимости от факта отмены), откажутся от услуг

```python
def profit_pred(row, col, refuse=False):
    if refuse:
        mod = 0.5
    else:
        mod = 1
    if (row[col]==0 and row['is_canceled']==0):
        income = row['price_per_night']*row['multiplier']*(row['stays_in_week_nights']+row['stays_in_weekend_nights'])
        expanse = row['room_service']* (1 + (row['stays_in_week_nights']+row['stays_in_weekend_nights']) // 2)
    elif (row[col]==1 and row['is_canceled']==0):
        income = row['price_per_night']*row['multiplier']*(row['stays_in_week_nights']+row['stays_in_weekend_nights']) * mod
        expanse = row['room_service']* (1 + (row['stays_in_week_nights']+row['stays_in_weekend_nights']) // 2) * mod
    elif (row[col]==0) and (row['is_canceled']==1):
        income = 0
        expanse = row['room_service'] + row['price_per_night']*row['multiplier']
    elif (row[col]==1) and (row['is_canceled']==1):
        income = (row['price_per_night']*row['multiplier']*0.8 + row['room_service']*0.8) * mod
        expanse = (row['room_service'] + row['price_per_night']*row['multiplier']) * mod
    profit_pred = income - expanse
    return profit_pred
```

Сделаем график кросс-валидации чтобы понять, что наши результаты стабильны и не случайны.<br />

```python
def cross_val_plot(model, features, target):
    plt.plot(cross_val_score(model, features, target, cv=5, scoring='recall', n_jobs=-1))
    plt.title('Показатели Recall по блокам кроссвалидации')
    plt.xlabel('Блок')
    plt.ylabel('Метрика Recall')
    plt.xticks([0, 1, 2, 3, 4, 5])
    plt.show()
```
